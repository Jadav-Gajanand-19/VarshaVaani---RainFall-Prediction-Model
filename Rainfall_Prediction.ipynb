```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('District_Rainfall_Normal_0.csv')
```

#Data PreProcessing :
###1.Removing null values
###2.Removing duplicate values
###3.Removing outliers
###4.Label Encoding

There are no null values hence there is no need to remove null values

There are no duplicate values hence there is no need to remove duplicate values

```python
# REMOVING OUTLIERS

for col in df.columns:
  if df[col].dtype !='object':
    sns.boxplot(df[col])
    plt.xlabel(col)
    plt.show()
```

```python
for col in df.columns:
  if df[col].dtype !='object':
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    Iqr=q3-q1
    lower_limit=q1-1.5*Iqr
    upper_limit=q3+1.5*Iqr
    df[col]=np.where(df[col]>upper_limit,upper_limit,np.where(df[col]<lower_limit,lower_limit,df[col]))
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

for col in df.columns:
  if df[col].dtype=="object":
    df[col]=le.fit_transform(df[col])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
df=pd.DataFrame(ss.fit_transform(df),columns=df.columns)
```

#PreProcessing is Done

#Lets Move towards Machine Learning Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

ml=LinearRegression()
X=df.drop('ANNUAL',axis=1)
Y=df['ANNUAL']
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
ml.fit(x_train,y_train)
y_pred=ml.predict(x_test)
acc=r2_score(y_test,y_pred)
print(round(acc*100,2))
```

```python
# Assuming 'X' is a DataFrame and 'Y' is a Series
for column in X.columns:
    plt.figure(figsize=(8, 6))  # Adjust figure size as needed
    plt.scatter(X[column], Y, color='blue', label='Data points')
    plt.plot(X[column], ml.predict(X), color='red', label='Regression line')
    plt.xlabel(column)  # Use column name for x-axis label
    plt.ylabel('Rainfall in (mm)')
    plt.title(f'Linear Regression - Rainfall Prediction ({column} vs. Annual)')  # Customize title
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import pickle

# Save the model
with open('Rainfall_Prediction_model.pkl', 'wb') as file:
    pickle.dump(ml, file)
```
